```python
import sys
sys.path.append('../')
from common import COUNTRYCODE2CONT, COUNTRYCODE2REGION, REGIONS, COUNTRIES, COUNTRYCODE2NAME
import pandas as pd
import numpy as np
import seaborn as sns
import time
import plotly
with open('/Users/timlee/Dropbox/keys/plotly_apikey.txt','r') as f:
    api_key = f.read()
plotly.tools.set_credentials_file(username='tdlee', api_key=api_key)

import plotly.plotly as ply
import plotly.graph_objs as go
from plotly.grid_objs import Grid, Column
import pandas as pd

import plotly.plotly as py
from plotly.grid_objs import Grid, Column
from plotly.tools import FigureFactory as FF 

import pandas as pd
import time


TARGET_IND =[
    'Export volume index (2000 = 100)',
    'Export value index (2000 = 100)',
    'Import volume index (2000 = 100)',
    'Import value index (2000 = 100)',
    'GDP per capita (current US$)',
    'Population, total'
]

TARGET_CODE = [
    'BM.GSR.GNFS.CD',
    'BX.GSR.GNFS.CD',
    'TM.VAL.MRCH.AL.ZS',
    'TX.VAL.MRCH.AL.ZS',
    'TM.VAL.MRCH.CD.WT',
    'TX.VAL.MRCH.CD.WT',
    'NY.GDP.PCAP.CD',
    'SP.POP.TOTL'
]

RENAME_COLS = {
    'BM.GSR.GNFS.CD': 'importval',
    'BX.GSR.GNFS.CD': 'exportval',
    'TM.VAL.MRCH.AL.ZS': 'importmerchpct',
    'TX.VAL.MRCH.AL.ZS': 'exportmerchpct',
    'TM.VAL.MRCH.CD.WT': 'importmerchtot',
    'TX.VAL.MRCH.CD.WT': 'exportmerchtot',
    'NY.GDP.PCAP.CD': 'gdpPercap',
    'SP.POP.TOTL':'pop',
    'CountryCode': 'countrycode',
    'country':'country',
    'continent': 'continent',
    'Year': 'year'
}

TARGET_IND=[
'Imports of goods and services (BoP, current US$)',
'Exports of goods and services (BoP, current US$)',
'Merchandise imports from high-income economies (% of total merchandise imports)'
'Merchandise exports to high-income economies (% of total merchandise exports)',
'GDP per capita (current US$)',
'Population, total'
]

CUSTOM_COLORS = {
    'Asia': 'rgb(228,26,28)',
    'Europe': 'rgb(55,126,184)',
    'Africa': 'rgb(77,175,74)',
    'Americas': 'rgb(152,78,163)',
    'Oceania': 'rgb(255,127,0)'
}
```

```python
def import_data():
    df = pd.read_csv('/Users/timlee/data/dataviz/kaggle/indicators.csv', low_memory=False)
    df.dropna(inplace=True)
    df.sort_values(['CountryCode','Year'], inplace=True)
    
    # excluding world bank
    df = df[df['CountryCode']!= 'WBG'].copy()
    
    # excluding the region codes
    df = df[df['CountryCode'].isin(REGION_CODES)==False]
    
    # missing data in 2014
    df = df[(df['Year'] < 2014) & (df['Year'] > 2004)].copy()
    
    return df

def transform_import_export(df, checkfields):
    mask = df['IndicatorCode'].isin(TARGET_CODE)
    pvt_tbl = df[mask].pivot_table(index=['CountryCode','Year'], columns='IndicatorCode', values='Value', aggfunc=np.sum ).reset_index()
    
    # drop missing values
    drop_ccs = []
    for fl in checkfields:
        drop_ccs += list(pvt_tbl.loc[pvt_tbl[fl].isna(), 'CountryCode'].unique())
    drop_ccs += list(pvt_tbl.loc[pvt_tbl['NY.GDP.PCAP.CD'].isna(), 'CountryCode'].unique())
    
    pvt_tbl = pvt_tbl[pvt_tbl['CountryCode'].isin(drop_ccs)==False].copy()    
    pvt_tbl['country'] = pvt_tbl['CountryCode'].map(COUNTRYCODE2NAME)
    pvt_tbl['continent'] = pvt_tbl['CountryCode'].map(COUNTRYCODE2CONT)

#     pvt_tbl.columns = ['countrycode', 'year','gdpPercap','pop',
#                        'importvol', 'importval', 'exportvol', 'exportval',
#                        'country', 'continent' ]

    pvt_tbl.columns = [RENAME_COLS[col] for col in pvt_tbl.columns] 
    
    pvt_tbl['importmerchval'] = pvt_tbl['importmerchpct'] * pvt_tbl['importmerchtot']
    pvt_tbl['exportmerchval'] = pvt_tbl['exportmerchpct'] * pvt_tbl['exportmerchtot']
    return pvt_tbl

def makegrid(df):
    dataset = df
    years_from_col = set(dataset['year'])
    years_ints = sorted(list(years_from_col))
    years = [str(year) for year in years_ints]

    # make list of continents
    continents = []
    for continent in dataset['continent']:
        if continent not in continents: 
            continents.append(continent)

    columns = []
    column_names = []
    # make grid
    for year in years:
        for continent in continents:
            dataset_by_year = dataset[dataset['year'] == int(year)]
            dataset_by_year_and_cont = dataset_by_year[dataset_by_year['continent'] == continent]
            for col_name in dataset_by_year_and_cont:
                # each column name is unique
                column_name = '{year}_{continent}_{header}_gapminder_grid'.format(
                    year=year, continent=continent, header=col_name
                )
                column_names.append(column_name)
                a_column = Column(list(dataset_by_year_and_cont[col_name]), column_name)
                columns.append(a_column)

    # upload grid
    grid = Grid(columns)
    url = py.grid_ops.upload(grid, 'gapminder_grid'+str(time.time()), auto_open=False)
    return grid, continents, years, column_names

def initialize_fig(rng=[6, 13]):
    figure = {
        'data': [],
        'layout': {},
        'frames': [],
        'config': {'scrollzoom': True}
    }

    # fill in most of layout
    figure['layout']['xaxis'] = {'range': rng, 'title': 'Exports', 'type': 'log', 'gridcolor': '#FFFFFF'}
    figure['layout']['yaxis'] = {'range': rng, 'title': 'Imports', 'type': 'log', 'gridcolor': '#FFFFFF'}
    figure['layout']['width'] = 800
    figure['layout']['height'] = 800
    figure['layout']['hovermode'] = 'closest'
    figure['layout']['plot_bgcolor'] = 'rgb(223, 232, 243)'
    
    figure['layout']['updatemenus'] = [
        {
            'buttons': [
                {
                    'args': [None, {'frame': {'duration': 1000, 'redraw': False},
                             'fromcurrent': True, 'transition': {'duration': 300, 'easing': 'quadratic-in-out'}}],
                    'label': 'Play',
                    'method': 'animate'
                },
                {
                    'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                    'transition': {'duration': 0}}],
                    'label': 'Pause',
                    'method': 'animate'
                }
            ],
            'direction': 'left',
            'pad': {'r': 10, 't': 87},
            'showactive': False,
            'type': 'buttons',
            'x': 0.1,
            'xanchor': 'right',
            'y': 0,
            'yanchor': 'top'
        }
    ]
    return figure

def get_slider():
    sliders_dict = {
        'active': 0,
        'yanchor': 'top',
        'xanchor': 'left',
        'currentvalue': {
            'font': {'size': 20},
            'prefix': 'Year:',
            'visible': True,
            'xanchor': 'right'
        },
        'transition': {'duration': 300, 'easing': 'cubic-in-out'},
        'pad': {'b': 10, 't': 50},
        'len': 0.9,
        'x': 0.1,
        'y': 0,
        'steps': []
    }
    return sliders_dict

def apply_data_to_fig(grid, 
                      figure,
                      sliders_dict,
                      continents,
                      years,
                      column_names,
                      start_year=2005,
                      xflname='exportval',
                      yflname='importval',
                      hovername='country',
                      szflname='gdpPercap',
                      szref=1):
    year = start_year
    xfield = xflname
    yfield = yflname
    hover = hovername
    szfield = szflname
    szref = szref
    col_name_template = '{year}_{continent}_{header}_gapminder_grid'
    for continent in continents:
        x_col = col_name_template.format(
                year=year, continent=continent, header=xfield
            )

        y_col = col_name_template.format(
                year=year, continent=continent, header=yfield
                )

        sz_col = col_name_template.format(
                    year=year, continent=continent, header=szfield
                )
        hover_col = col_name_template.format(
                    year=year, continent=continent, header=hover
                )
        
        if x_col not in column_names:
            print(x_col)
        if y_col not in column_names:
            print(y_col)
        if sz_col not in column_names:
            print(sz_col)
        if hover_col not in column_names:
            print(hover_col)

        data_dict = {
            'xsrc': grid.get_column_reference(x_col),
            'ysrc': grid.get_column_reference(y_col),
            'mode': 'markers',
            'textsrc': grid.get_column_reference(col_name_template.format(
                year=year, continent=continent, header=hover
            )),
            'marker': {
                'sizemode': 'area',
                'sizeref': szref,
                'sizesrc': grid.get_column_reference(col_name_template.format(
                     year=year, continent=continent, header=szfield
                )),
                'color': CUSTOM_COLORS[continent]
            },
            'name': continent
        }
        figure['data'].append(data_dict)
    
    
    for year in years:
        frame = {'data': [], 'name': str(year)}
        for continent in continents:
            x_col = col_name_template.format(
                    year=year, continent=continent, header=xfield
                )

            y_col = col_name_template.format(
                    year=year, continent=continent, header=yfield
                    )

            sz_col = col_name_template.format(
                        year=year, continent=continent, header=szfield
                    )
            hover_col = col_name_template.format(
                        year=year, continent=continent, header=hover
                    )

            if x_col not in column_names:
                print(x_col)
            if y_col not in column_names:
                print(y_col)
            if sz_col not in column_names:
                print(sz_col)
            if hover_col not in column_names:
                print(hover_col)

            data_dict = {
                'xsrc': grid.get_column_reference(x_col),
                'ysrc': grid.get_column_reference(y_col),
                'mode': 'markers',
                'textsrc': grid.get_column_reference(hover_col),
                'marker': {
                    'sizemode': 'area',
                    'sizeref': szref,
                    'sizesrc': grid.get_column_reference(sz_col),
                    'color': CUSTOM_COLORS[continent]
                },
                'name': continent
            }
            frame['data'].append(data_dict)

        figure['frames'].append(frame)
        slider_step = {'args': [
            [year],
            {'frame': {'duration': 300, 'redraw': False},
             'mode': 'immediate',
           'transition': {'duration': 300}}
         ],
         'label': year,
         'method': 'animate'}
        sliders_dict['steps'].append(slider_step)

    figure['layout']['sliders'] = [sliders_dict]
```

```python
df = import_data()
checkfields = ['BM.GSR.GNFS.CD', 'TM.VAL.MRCH.CD.WT', 'TX.VAL.MRCH.CD.WT']
df_pvt = transform_import_export(df, checkfields)
```

```python
df_pvt['gdpPercap']  = np.sqrt(df_pvt['gdpPercap'].astype(int))
grid, continents, years, column_names = makegrid(df_pvt)
fig = initialize_fig(rng=[3,16])
sliders_dict = get_slider()
apply_data_to_fig(grid,
                  fig,
                  sliders_dict,
                  continents,
                  years,
                  column_names,
                  xflname='exportmerchval',
                  yflname='importmerchval',
                  szref=0.1)
```

```python
ply.icreate_animations(fig, 'gapminder_example'+str(time.time()))
```
